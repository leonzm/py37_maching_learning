## 1.1 ÂÆö‰πâÁΩëÁªú
![](https://fancyerii.github.io/img/pytorch/mnist.png)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):

	def __init__(self):
		super(Net, self).__init__()
		# ËæìÂÖ•ÊòØ1‰∏™ÈÄöÈÅìÁöÑÁÅ∞Â∫¶ÂõæÔºåËæìÂá∫6‰∏™ÈÄöÈÅì(feature map)Ôºå‰ΩøÁî®5x5ÁöÑÂç∑ÁßØÊ†∏
		self.conv1 = nn.Conv2d(1, 6, 5)
		# Á¨¨‰∫å‰∏™Âç∑ÁßØÂ±Ç‰πüÊòØ5x5ÔºåÊúâ16‰∏™ÈÄöÈÅì
		self.conv2 = nn.Conv2d(6, 16, 5)
		# ÂÖ®ËøûÊé•Â±Ç
		self.fc1 = nn.Linear(16 * 5 * 5, 120)
		self.fc2 = nn.Linear(120, 84)
		self.fc3 = nn.Linear(84, 10)
	
	def forward(self, x):
		# 32x32 -> 28x28 -> 14x14 
		x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
		# 14x14 -> 10x10 -> 5x5
		x = F.max_pool2d(F.relu(self.conv2(x)), 2)
		x = x.view(-1, self.num_flat_features(x))
		x = F.relu(self.fc1(x))
		x = F.relu(self.fc2(x))
		x = self.fc3(x)
		return x
	
	def num_flat_features(self, x):
		size = x.size()[1:]  # Èô§‰∫ÜbatchÁª¥Â∫¶‰πãÂ§ñÁöÑÂÖ∂ÂÆÉÁª¥Â∫¶„ÄÇ
		num_features = 1
		for s in size:
		    num_features *= s
		return num_features


net = Net()
print(net)
```

```python
params = list(net.parameters())  # ÊâÄÊúâÂ±ÇÂèÇÊï∞
print(len(params))
print(params[0].size())  # conv1 Â±ÇÂèÇÊï∞
```

## 1.2 ÊµãËØïÁΩëÁªú

```python
input = torch.randn(1, 1, 32, 32)  # ÊâπÊ¨°„ÄÅÈÄöÈÅì„ÄÅÈ´ò„ÄÅÂÆΩ
out = net(input)
print(out)
# tensor([[-0.1407, -0.1570,  0.0113, -0.1059,  0.1226,  0.0791, -0.0023,  0.0688, -0.0138,  0.0706]], grad_fn=<AddmmBackward0>)

# ÈªòËÆ§ÁöÑÊ¢ØÂ∫¶‰ºöÁ¥ØÂä†ÔºåÂõ†Ê≠§ÈÄöÂ∏∏Âú®backward‰πãÂâçÊ∏ÖÈô§Êéâ‰πãÂâçÁöÑÊ¢ØÂ∫¶ÂÄº
net.zero_grad()
out.backward(torch.randn(1, 10))
```

## 1.3 ÊçüÂ§±ÂáΩÊï∞

```python
output = net(input)
target = torch.arange(1, 11, dtype=torch.float)  # Èöè‰æø‰º™ÈÄ†ÁöÑ‰∏Ä‰∏™‚ÄúÁúüÂÆûÂÄº‚Äù 
target = target.view(1, -1)  # ÊääÂÆÉÂèòÊàêoutputÁöÑshape(1, 10) 
criterion = nn.MSELoss()

loss = criterion(output, target)
print(loss)
```

```python
print(loss.grad_fn)  # MSELoss
print(loss.grad_fn.next_functions[0][0])  # Add
print(loss.grad_fn.next_functions[0][0].next_functions[0][0])  # Expand
```

## 1.4 ËÆ°ÁÆóÊ¢ØÂ∫¶

```python
net.zero_grad()     # Ê∏ÖÊéâtensorÈáåÁºìÂ≠òÁöÑÊ¢ØÂ∫¶ÂÄº„ÄÇ

print('conv1.bias.grad before backward')
print(net.conv1.bias.grad)  # None

loss.backward()

print('conv1.bias.grad after backward')
print(net.conv1.bias.grad)  # tensor([-0.0412,  0.0738,  0.0792, -0.1728,  0.0262,  0.0698])
```

## 1.5 Êõ¥Êñ∞ÂèÇÊï∞
> Êõ¥Êñ∞ÂèÇÊï∞ÊúÄÁÆÄÂçïÁöÑÊñπÊ≥ïÊòØ‰ΩøÁî®ÈöèÊú∫Ê¢ØÂ∫¶‰∏ãÈôç(SGD)Ôºö ùë§ùëíùëñùëî‚Ñéùë°=ùë§ùëíùëñùëî‚Ñéùë°‚àíùëôùëíùëéùëüùëõùëñùëõùëîùëüùëéùë°ùëí‚àóùëîùëüùëéùëëùëñùëíùëõùë°

```python
learning_rate = 0.01
for f in net.parameters():
	f.data.sub_(f.grad.data * learning_rate)
```

```python
# ‰ΩøÁî®Êõ¥Â§çÊùÇÁöÑ‰ºòÂåñÊñπÊ≥ïÔºåÊØîÂ¶ÇSGD, Nesterov-SGD, Adam, RMSPropÁ≠âÁ≠â.
import torch.optim as optim

# ÂàõÂª∫optimizerÔºåÈúÄË¶Å‰º†ÂÖ•ÂèÇÊï∞Âíålearning rate
optimizer = optim.SGD(net.parameters(), lr=0.01)

# Ê∏ÖÈô§Ê¢ØÂ∫¶
optimizer.zero_grad()  
output = net(input)
loss = criterion(output, target)
loss.backward()
optimizer.step()    # optimizer‰ºöËá™Âä®Â∏ÆÊàë‰ª¨Êõ¥Êñ∞ÂèÇÊï∞
```

## 2.3 Êï∞ÊçÆÂ§ÑÁêÜ

```python
# ÈÄöËøá‰ΩøÁî® torchvisionÔºåÂèØ‰ª•ËΩªÊùæÁöÑÂä†ËΩΩ CIFAR10 Êï∞ÊçÆÈõÜ
import torch
import torchvision
import torchvision.transforms as transforms
```

```python
# torchvision ËØªÂèñÁöÑ datasets ÊòØ PILImage ÂØπË±°ÔºåÂÆÉÁöÑÂèñÂÄºËåÉÂõ¥ÊòØ [0, 1]ÔºåÊääÂÆÉËΩ¨Êç¢Âà∞ËåÉÂõ¥[-1, 1]
transform = transforms.Compose([
    transforms.ToTensor(),  # ËΩ¨Êç¢‰∏∫Âº†Èáè
	transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Ê†áÂáÜÂåñ„ÄÇ‰∏§‰∏™ÂèÇÊï∞Ôºömean Âíå std
])

trainset = torchvision.datasets.CIFAR10(root='/tmp/to/data/train', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='/tmp/to/data/test', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4, shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
# Êù•ÁúãÂá†Âº†ÂõæÁâá
import numpy as np
import matplotlib.pyplot as plt


def imshow(img):
    """
    # ÊòæÁ§∫ÂõæÁâáÁöÑÂáΩÊï∞
    """
    img = img / 2 + 0.5  # [-1,1] -> [0,1]
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))  # (channel, width, height) -> (width, height, channel)

# ÈöèÊú∫ÈÄâÊã©‰∏Ä‰∫õÂõæÁâá
dataiter = iter(trainloader)
# images, labels = dataiter.next()
images, labels = next(dataiter)

# ÊòæÁ§∫ÂõæÁâá
imshow(torchvision.utils.make_grid(images))
# ÊâìÂç∞label
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

## 2.4 ÂÆö‰πâÂç∑ÁßØÁΩëÁªú

```python
# ÂÆö‰πâ‰∏§‰∏™Âç∑ÁßØÂ±ÇÁöÑÁΩëÁªú
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
	def __init__(self):
		super(Net, self).__init__()
		self.conv1 = nn.Conv2d(3, 6, 5)
		self.pool = nn.MaxPool2d(2, 2)
		self.conv2 = nn.Conv2d(6, 16, 5)
		self.fc1 = nn.Linear(16 * 5 * 5, 120)
		self.fc2 = nn.Linear(120, 84)
		self.fc3 = nn.Linear(84, 10)
	
	def forward(self, x):
		x = self.pool(F.relu(self.conv1(x)))
		x = self.pool(F.relu(self.conv2(x)))
		x = x.view(-1, 16 * 5 * 5)
		x = F.relu(self.fc1(x))
		x = F.relu(self.fc2(x))
		x = self.fc3(x)
		return x

net = Net()
```

```python
# ÂÆö‰πâÊçüÂ§±ÂáΩÊï∞Âíå optimizer„ÄÇËøôÈáå‰ΩøÁî®‰∫§ÂèâÁÜµÊçüÂ§±ÂáΩÊï∞ÔºåOptimizer ‰ΩøÁî®Â∏¶ÂÜ≤ÈáèÁöÑ SGD
import torch.optim as optim

criterion = nn.CrossEntropyLoss()  # ‰∫§ÂèâÁÜµÊçüÂ§±ÂáΩÊï∞
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)  # ‰ºòÂåñÂô®
```

```python
# ËÆ≠ÁªÉÁΩëÁªú
for epoch in range(2):  # ËøôÈáåÂè™Ëø≠‰ª£ 2 ‰∏™ epochÔºåÂÆûÈôÖÂ∫îËØ•ËøõË°åÊõ¥Â§öÊ¨°ËÆ≠ÁªÉ 
	running_loss = 0.0
	for i, data in enumerate(trainloader, 0):
		# ÂæóÂà∞ËæìÂÖ•
		inputs, labels = data
        
		# Ê¢ØÂ∫¶Ê∏ÖÈõ∂ 
		optimizer.zero_grad()
        
		# forward + backward + optimize
		outputs = net(inputs)  # Ê≠£Âêë‰º†Êí≠ÔºåËÆ°ÁÆóËæìÂá∫ÂÄº
		loss = criterion(outputs, labels)  # ËÆ°ÁÆóÊçüÂ§±
		loss.backward()  # ÂèçÂêë‰º†Êí≠ÔºåËÆ°ÁÆóÊ¢ØÂ∫¶
		optimizer.step()  # Êõ¥Êñ∞ÂèÇÊï∞
		
		# ÂÆö‰πâÁªüËÆ°‰ø°ÊÅØ„ÄÇÊØè‰∏§ÂçÉÊ¨°Ëø≠‰ª£ËÆ°ÁÆó‰∏ÄÊ¨°ÊçüÂ§±
		running_loss += loss.item()
		if i % 2000 == 1999:
			print('[%d, %5d] loss: %.6f' % (epoch + 1, i + 1, running_loss / 2000))
		running_loss = 0.0

print('Finished Training')
```

## 2.5 Âú®ÊµãËØïÊï∞ÊçÆÈõÜ‰∏äËøõË°åËÆ≠ÁªÉ

```python
# ÂèØ‰ª•‰ΩøÁî®ÊµãËØïÊï∞ÊçÆÈõÜ‰∏äÁöÑÊï∞ÊçÆÊù•ËøõË°åÊµãËØï„ÄÇÈ¶ñÂÖàÈöèÊú∫ÊäΩÂèñÂá†‰∏™Ê†∑Êú¨Êù•ËøõË°åÊµãËØï
dataiter = iter(testloader)
images, labels = next(dataiter)

imshow(torchvision.utils.make_grid(images))
print('GroundTruth: ', ' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

```python
# Áî®Ê®°ÂûãÊù•È¢ÑÊµã‰∏Ä‰∏ãÔºåÁúãÁúãÊòØÂê¶Ê≠£Á°ÆÈ¢ÑÊµãÔºö
outputs = net(images)

# outputsÊòØ10‰∏™ÂàÜÁ±ªÁöÑlogits„ÄÇÂú®ËÆ≠ÁªÉÁöÑÊó∂ÂÄôÈúÄË¶ÅÁî®softmaxÊääÂÆÉÂèòÊàêÊ¶ÇÁéá(CrossEntropyLossÂ∏ÆÊàë‰ª¨ÂÅö‰∫Ü)Ôºå‰ΩÜÊòØÈ¢ÑÊµãÁöÑÊó∂ÂÄôÊ≤°ÊúâÂøÖË¶ÅÔºåÂõ†‰∏∫Âè™ÈúÄË¶ÅÁü•ÈÅìÂì™‰∏™ÂàÜÁ±ªÁöÑÊ¶ÇÁéáÂ§ßÂ∞±Ë°å
_, predicted = torch.max(outputs, 1)
print('Predicted: ', ' '.join('%5s' % classes[predicted[j]] for j in range(4)))
```

```python
# Âú®Êï¥‰∏™ÊµãËØïÈõÜÂêà‰∏äÁöÑÊïàÊûú
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
	    images, labels = data
	    outputs = net(images)
	    _, predicted = torch.max(outputs.data, 1)
	    total += labels.size(0)
	    correct += (predicted == labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (100 * correct / total))
```

```python
# ÊØè‰∏™ÂàÜÁ±ªÁöÑÂáÜÁ°ÆÁéá
class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))
with torch.no_grad():
	for data in testloader:
		images, labels = data
		outputs = net(images)
		_, predicted = torch.max(outputs, 1)
		c = (predicted == labels).squeeze()
		for i in range(4):
			label = labels[i]
			class_correct[label] += c[i].item()
			class_total[label] += 1


for i in range(10):
	print('Accuracy of %5s : %2d %%' % (classes[i], 100 * class_correct[i] / class_total[i]))
```

## 2.6 GPU ‰∏äËÆ≠ÁªÉ

```python
# ÁúãÁúãÊòØÂê¶Êúâ GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
# Áî® GPU ËøõË°åËÆ≠ÁªÉ
class Net2(nn.Module):
    def __init__(self):
        super(Net2, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5).to(device)
        self.pool = nn.MaxPool2d(2, 2).to(device)
        self.conv2 = nn.Conv2d(6, 16, 5).to(device)
        self.fc1 = nn.Linear(16 * 5 * 5, 120).to(device)
        self.fc2 = nn.Linear(120, 84).to(device)
        self.fc3 = nn.Linear(84, 10).to(device)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net2()
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)

for epoch in range(20):
	running_loss = 0.0
	for i, data in enumerate(trainloader, 0):
		# ÂæóÂà∞ËæìÂÖ•
		inputs, labels = data 
		inputs, labels = inputs.to(device), labels.to(device) 
		# Ê¢ØÂ∫¶Ê∏ÖÈõ∂ 
		optimizer.zero_grad()
		
		# forward + backward + optimize
		outputs = net(inputs)
		loss = criterion(outputs, labels)
		loss.backward()
		optimizer.step()
		
		# ÂÆö‰πâÁªüËÆ°‰ø°ÊÅØ
		running_loss += loss.item()
		if i % 2000 == 1999:
			print('[%d, %5d] loss: %.6f' % (epoch + 1, i + 1, running_loss / 2000))
			running_loss = 0.0
		print('Finished Training')
```
